```python
# Matplolib - low level graph plotting lib ; serves as a visualization utility
# open source & free

# mostly written in python; little in C, Objective-C & Javascript for plateform compatibility
```

```python
import matplotlib.pyplot as plt 
import numpy as np
```

```python
# drawing a line 

xpoints = np.array([0, 6])
ypoints = np.array([0, 250])

plt.plot(xpoints, ypoints)
plt.show()


# drawing markers/ points
xpoints = np.array([0, 6])
ypoints = np.array([0, 250])

plt.plot(xpoints, ypoints, "o")
plt.show()


# multiple points
xpoints = np.array([1, 4, 6, 7])
ypoints = np.array([100, 50, 40, 250])

plt.plot(xpoints, ypoints,)
plt.show()


# default X-points; x points will be taken as 0.0, 0.5, 1.0, 1.5 if not given
ypoints = np.array([100, 50, 40, 250])

plt.plot(ypoints)
plt.show()
```

```python
# markers
xpoints = np.array([1, 4, 6, 7])
ypoints = np.array([100, 50, 40, 250])

plt.plot(xpoints, ypoints, marker = 'p')
plt.show()

# fmt parameter (marker|line|color)
xpoints = np.array([1, 4, 6, 7])
ypoints = np.array([100, 50, 40, 250])

plt.plot(xpoints, ypoints, 'o:r') # 'o:r' : shortcut string notation parameter
plt.show()

"""
'-'	    Solid line	
':'   	Dotted line	
'--'	Dashed line	
'-.'	Dashed/dotted line
"""
# markersize or ms and markeredgecolor or mec (to color on edge of marker) and markerfacecolor or mfc (to color inside the edge of the markers) 
xpoints = np.array([1, 4, 6, 7])
ypoints = np.array([100, 50, 40, 250])

plt.plot(xpoints, ypoints, marker = 'o', ms = 20, mec = 'r', mfc  = 'r') # 'o:r' : shortcut string notation parameter; use both ms & mfc
plt.show()
```
