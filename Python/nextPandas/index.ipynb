```python
import pandas as pd

mydataset = {
    "cars": ["BMW", "Volvo", "Food"],
    "passings": [3, 7, 2]
}


myvar = pd.DataFrame(mydataset)

print(myvar)
print(pd.__version__)
```

```python
label = [1, 7, 2]
col1 = [1, 7, 2]
col2 = [1, 7, 2]

myvar = pd.Series(col1, index = ["x", "y", "z"])
myvar1 = pd.Series(col1, label)
myvar2 = pd.Series(col2) # no label => 0, 1, 2 indexing

print(myvar)
print(myvar1)
print(myvar2)
print(myvar2[1])
```

```python
calories = {"day1": 420, "day2": 380, "day3": 390}

myvar = pd.Series(calories) # keys => label

print(myvar)

# selecting some items
myvar = pd.Series(calories, index = ["day1", "day2"]) 

print(myvar)
```

```python
data = {
  "calories": [420, 380, 390],
  "duration": [50, 40, 45]
}

#load data into a DataFrame object:
df = pd.DataFrame(data)

# print(df)

# locate row 
print(df.loc[0]) # returns series

print(df.loc[[0, 1]]) # returns dataframe
```

```python
data = {
  "calories": [420, 380, 390],
  "duration": [50, 40, 45]
}

df = pd.DataFrame(data, index = ["day1", "day2", "day3"])

print(df)
print(df.loc["day2"])
```
